```python
import requests
from bs4 import BeautifulSoup

def search_drug(ingrEngName):
    url = "https://nedrug.mfds.go.kr/searchDrug"
    params = {
        "sort": "",
        "sortOrder": "false",
        "searchYn": "true",
        "ExcelRowdata": "",
        "page": "1",
        "searchDivision": "detail",
        "itemName": "",
        "itemEngName": "",
        "entpName": "",
        "entpEngName": "",
        "ingrName1": "",
        "ingrName2": "",
        "ingrName3": "",
        "ingrEngName": ingrEngName,
        "itemSeq": "",
        "stdrCodeName": "",
        "atcCodeName": "",
        "indutyClassCode": "",
        "sClassNo": "",
        "narcoticKindCode": "",
        "cancelCode": "",
        "etcOtcCode": "",
        "makeMaterialGb": "",
        "searchConEe": "AND",
        "eeDocData": "",
        "searchConUd": "AND",
        "udDocData": "",
        "searchConNb": "AND",
        "nbDocData": "",
        "startPermitDate": "",
        "endPermitDate": ""
    }
    response = requests.post(url, data=params)
    soup = BeautifulSoup(response.content, 'html.parser')
    # 테이블의 내용을 찾아서 가져옵니다.
    table = soup.find('table', class_='table_scroll')
    if table:
        rows = table.find_all('tr')
        # 첫 번째 행은 헤더입니다. 그러므로 제외합니다.
        for row in rows[1:]:
            # 각 행의 데이터를 가져옵니다.
            columns = row.find_all('td')
            # 각 열의 데이터를 출력하거나 원하는 방식으로 처리합니다.
            for column in columns:
                print(column.get_text().strip())
            print('------')  # 각 행의 구분을 위해 구분선 출력
    else:
        print("테이블이 없습니다.")

# 특정 검색어를 입력하여 검색을 수행합니다.
search_term = input("검색어를 입력하세요: ")
search_drug(search_term)
```
