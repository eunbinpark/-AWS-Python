```python
csv_data = []
with open('lott.CSV') as f:
    for line in f:
        csv_data.append(line[:-1].split(','))
        #csv_data.append(line.replace('\n', '').split(','))
print(csv_data[:3])

int(csv_data[0][-1])
```

```python
import math

def mean(num_list):
    return sum(num_list) / len(num_list)

def median(num_list):
    num_list.sort()
    if len(num_list) % 2 == 1:
        i = len(num_list) // 2
        return num_list[i]
    else:
        i = len(num_list) // 2
        return (num_list[i] + num_list[i - 1]) / 2

def dev(num_list): # 편차
    m = mean(num_list)
    return [x - m for x in num_list]

def var(num_list): # 분산
    n = len(num_list)
    d = dev(num_list)
    return sum([x * x for x in d]) / (n - 1)

def stdev(num_list): # 표준편차
    return math.sqrt(var(num_list))
```

```python
# 1. 전체 회차를 통틀어 빈도수 (mode) : 숫자별 카운트 (몇번이나 나왔나)
# 2. 회차별 - 평균값, 표준편차

count_list = []

for i in csv_data:
    count_list = count_list + i
count_set = set(count_list)

for i in count_set:
    print('count of', i, ':', count_list.count(i))
    
for e, i in enumerate(csv_data):
    print('average of', (e + 1), ' : ', mean(i))

for e, i in enumerate(csv_data):
    print('Standard Deviation of', (e + 1), ' : ', stdev(i))
```
