```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

url = 'https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt'
pd_data = pd.read_csv(url, sep = '\t')
print(pd_data)
```

```python
_, axe = plt.subplots()
#axe.plot(pd_data['BMI'])
axe.hist(pd_data['BP'], bins = 10, edgecolor = 'pink', linewidth = 2.0)
```

```python
_, axe = plt.subplots()
axe.boxplot(pd_data['AGE'])
```

```python
_, axe = plt.subplots()
axe.pie(pd_data['SEX'].value_counts())
```
