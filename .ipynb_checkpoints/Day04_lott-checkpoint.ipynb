```python
def mean(num_list):
    return sum(num_list) / len(num_list)

def mean_2(num_list):
    sum_value = 0
    for i in num_list:
        sum_value = sum_value + i
    return sum_value / len(num_list)

def median(num_list):
    num_list.sort()
    if len(num_list) % 2 == 1:
        i = len(num_list) // 2
        return num_list[i]
    else:
        i = len(num_list) // 2
        return (num_list[i] + num_list[i - 1]) / 2

def dev(num_list): # 편차
    m = mean(num_list)
    return [x - m for x in num_list]

def var(num_list):
    n = len(num_list)
    d = dev(num_list)
    return sum([x * x for x in d]) / (n - 1)

def stdev(num_list):
    return math.sqrt(var(num_list))

game1 = csv_data[0]
print(game1)
print(mean(game1))
print(mean_2(game1))
print(game1.sort())
print(game1[1:])

print(median([1, 5, 17, 26, 27, 35, 38]))
print(median([5, 17, 26, 27, 35, 38]))

# assert : 샘플테이터가 맞는지 확인
assert median([5, 17, 26, 27, 35, 38, 1]) == 26

print(dev([1, 5, 17, 26, 27, 35, 38]))
print(var([1, 5, 17, 26, 27, 35, 38]))
```

```python
# in-place
list_a = [5, 17, 26, 27, 35, 38, 1]
# 있는거 복사본 정렬 후 반환
print(sorted(list_a))
print('after sorted : ', list_a)

# 있는거 바로 정렬
list_a.sort()
print(list_a)
```

```python
# reference
list_a = [5, 17, 26, 27, 35, 38, 1]
list_b = list_a # shallow copy, 얕은 복사
print('list_a : ', list_a)
print('list_b : ', list_b)

list_a[0] = 100

print('list_a : ', list_a)
print('list_b : ', list_b)
```

```python
list_a = [5, 17, 26, 27, 35, 38, 1]
list_b = []

# deep copy
for i in list_a:
    list_b.append(i)

print('list_a : ', list_a)
print('list_b : ', list_b)

list_a[0] = 100

print('list_a : ', list_a)
print('list_b : ', list_b)
```
