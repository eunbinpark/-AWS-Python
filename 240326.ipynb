```python
import requests
from bs4 import BeautifulSoup

# 웹페이지 주소
url = "https://www.druginfo.co.kr/search2/search.aspx?q=%20sorafenib"

# HTML 가져오기
response = requests.get(url)
html = response.text

# BeautifulSoup을 이용해 HTML 파싱
soup = BeautifulSoup(html, "html.parser")

# subheader 클래스를 가진 div 요소 찾기
subheader_div = soup.find("div", class_="subheader")

# subheader_div가 존재하는지 확인
if subheader_div:
    # subheader_div의 하위에 있는 green-font 클래스를 가진 h2 태그 찾기
    green_font_h2 = subheader_div.find("h2", class_="green-font")

    # green-font 클래스를 가진 h2 태그가 존재하는지 확인
    if green_font_h2:
        # green-font 클래스를 가진 h2 태그의 부모 요소의 다음 sibling 요소 찾기
        res_table_res_div = green_font_h2.find_next_sibling("div", class_="res table-res")

        # 테이블 내용 추출
        if res_table_res_div:
            table = res_table_res_div.find("table")
            table_rows = table.find_all("tr")

            # 각 행의 셀 내용 출력
            for row in table_rows:
                cells = row.find_all("td")
                for cell in cells:
                    print(cell.text.strip(), end="\t")
                print()
    else:
        print("No green-font class found.")
else:
    print("No subheader class found.")
```

```python
import requests
from bs4 import BeautifulSoup

# 웹페이지 주소
url = "https://www.druginfo.co.kr/search2/search.aspx?q=%20sorafenib"

# HTML 가져오기
response = requests.get(url)
html = response.text

# BeautifulSoup을 이용해 HTML 파싱
soup = BeautifulSoup(html, "html.parser")

# "유통중인 제품"이 포함된 텍스트 찾기
target_text = "유통중인 제품"

# HTML에서 "유통중인 제품"이 포함된 태그 및 상위 태그 및 클래스 검사
for tag in soup.find_all(text=lambda text: text and target_text in text):
    # "유통중인 제품"을 포함하는 태그 찾기
    if target_text in tag:
        # 해당 태그의 상위 태그 및 클래스 출력
        print("텍스트 '{}'를 포함하는 태그 및 상위 태그 및 클래스:".format(target_text))
        parent = tag.find_parent()
        while parent:
            print(parent.name, parent.get("class"))
            parent = parent.find_parent()
        break
else:
    print("텍스트 '{}'를 포함하는 태그를 찾지 못했습니다.".format(target_text))
```

```python
import requests
from bs4 import BeautifulSoup

# URL 설정
url = "https://www.druginfo.co.kr/search2/search.aspx?q=%20sorafenib"

# 페이지 가져오기
response = requests.get(url)

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(response.text, 'html.parser')

# class가 "res table-res"인 요소 모두 찾기
target_elements = soup.find_all(class_="res table-res")

# 모든 요소에 대해 테이블 찾기
for target_element in target_elements:
    # 해당 요소 아래에 있는 테이블 찾기
    table = target_element.find('table')

    # 테이블이 존재하는 경우에만 처리
    if table:
        # 테이블에서 모든 행(tr) 찾기
        rows = table.find_all('tr')

        # 각 행에서 모든 셀(td) 값 추출하여 출력
        for row in rows:
            cells = row.find_all(['th', 'td'])  # th와 td 모두 포함하여 가져오기
            for cell in cells:
                print(cell.text.strip())  # 값 출력
            print('---')  # 각 행 구분을 위한 구분선 출력
```
